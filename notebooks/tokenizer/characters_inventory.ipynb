```python
from dotenv import load_dotenv
from pathlib import Path
from collections import Counter
import os
import sys
```

```python
def character_inventory(input_path: Path,output_path: Path):
    counter = Counter()

    files = list(input_path.glob("*.txt"))
    print(files)
    if not files:
        raise ValueError(f"No .txt files found in {input_path}")

    for file in files:
        with open(file, "r", encoding="utf-8") as f:
            for line in f:
                counter.update(line)
                
    total = sum(counter.values())
    unique = len(counter)

    print(f"Total characters: {total}")
    print(f"Unique characters: {unique}")
    with open(output_path, "w", encoding="utf-8") as f:
        for char, count in counter.most_common():
            f.write(f"{repr(char)}\t{count}\n")

    return counter
```

```python
load_dotenv()
project_root = Path(os.environ.get("PROJECT_ROOT", "."))
transcription_dir = project_root / "data"/"raw"
character_inventory_path = project_root / "tests"/"tokenizer/unique_characters_inventory.txt"
char_counts = character_inventory(input_path=transcription_dir,output_path=character_inventory_path)

#for char, count in char_counts.most_common():
#    print(f"{repr(char)} : {count}")
```
